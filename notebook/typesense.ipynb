```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('D:/trag/test.txt', encoding='utf-8')
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
docs = text_splitter.split_documents(documents)

embeddings = HuggingFaceEmbeddings()
```

```python
from langchain_community.vectorstores import Typesense

batch_size = 10
docsearch = None

for i in range(0, len(docs), batch_size):
    batch = docs[i:i+batch_size]
    
    print(f"Uploading batch {i//batch_size + 1}")
    
    if docsearch is None:
        docsearch = Typesense.from_documents(
            batch,
            embeddings,
            typesense_client_params={
                'host': 'y7s63wba8uei0dnjp-1.a2.typesense.net',
                'port': '443',
                'protocol': 'https',
                'typesense_api_key': 'EEFEMEhoZSea5eSSL9Gbe3abIvDy2vOie',
                'typesense_collection_name': 'test_collection',
                'connection_timeout_seconds': 20
            }
        )
    else:
        docsearch.add_documents(batch)
```
